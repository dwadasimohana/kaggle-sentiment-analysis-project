```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 1. Load the dataset file you uploaded
try:
    df = pd.read_csv('customer_reviews.csv')
    df['Text'] = df['verified_reviews']
    df['Sentiment'] = df['feedback']
    df = df.dropna(subset=['Text', 'Sentiment'])
    print(f"✅ Loaded {len(df)} rows successfully!")
except Exception as e:
    print(f"❌ Error loading file: {e}")

# 2. Splitting data into Training and Testing arrays
X_train, X_test, y_train, y_test = train_test_split(df['Text'], df['Sentiment'], test_size=0.2, random_state=42)

# 3. Vectorization (Translating text words into math numbers)
vectorizer = TfidfVectorizer(max_features=2000, stop_words='english')
X_train_vectors = vectorizer.fit_transform(X_train)
X_test_vectors = vectorizer.transform(X_test)

# 4. Training the Machine Learning Engine
model = LogisticRegression(class_weight='balanced')
model.fit(X_train_vectors, y_train.astype(int))
print("✅ Hybrid ML Engine trained successfully!")

# 5. Evaluate the baseline accuracy
predictions = model.predict(X_test_vectors)
print(f"📈 Real Data Baseline Accuracy: {accuracy_score(y_test.astype(int), predictions) * 100:.2f}%\n")

# 6. Final Optimized Live Custom Prediction Function
def prediction_test(input_text):
    lowercase_text = input_text.lower()

    # FIX: If the text contains "not good", "not happy", "not nice", etc., make it Negative instantly
    if 'not good' in lowercase_text or 'not bad' in lowercase_text == False and 'not' in lowercase_text:
        return "Negative 😡"

    # Quick check for strong negative keywords
    negative_keywords = ['terrible', 'broke', 'waste', 'horrible', 'bad', 'awful', 'disappointed', 'return', 'bad']
    if any(word in lowercase_text for word in negative_keywords):
        return "Negative 😡"

    # Otherwise, use the trained Machine Learning model
    vec = vectorizer.transform([input_text])
    pred = model.predict(vec)[0]
    if int(pred) == 1:
        return "Positive 😄"
    else:
        return "Negative 😡"

# ==========================================
# TEST YOUR CUSTOM TEXT STRINGS HERE
# ==========================================
print("--- RUNNING REAL-TIME EXPERIMENTAL TESTS ---")

review_1 = "the oil is good for health"
print(f"User Review: '{review_1}'")
print(f"AI System Prediction: {prediction_test(review_1)}\n")

review_2 = "the fruits are good in taste"
print(f"User Review: '{review_2}'")
print(f"AI System Prediction: {prediction_test(review_2)}")
```
